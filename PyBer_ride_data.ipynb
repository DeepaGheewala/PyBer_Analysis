```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Load in csv
pyber_ride_df = pd.read_csv("Resources/PyBer_ride_data.csv")
pyber_ride_df
```

```python
pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)")
plt.show()
```

```python
# Set x-axis and tick locations.
x_axis = np.arange(len(pyber_ride_df))
tick_locations = [value for value in x_axis]
# Plot the data.
pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)")
plt.xticks(tick_locations, pyber_ride_df["Month"])
plt.show()
```

```python
pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)", kind='bar')
plt.show()
```

```python
#x_axis = np.arange(len(pyber_ride_df))
#tick_locations = [value for value in x_axis]
#ax.set_xticks(tick_locations, pyber_ride_df["Month"])

from numpy.random import randn, rand

fig, ax= plt.subplots(figsize=(4, 4))
fig1, ax1= plt.subplots(figsize=(4, 4))

ax.set_yticks(np.arange(0, 51, step=5.0))
ax.set_title("Monthly Rider Data")

ax1.set_xticks(np.arange(0, 51, step=5.0))
ax1.set_title("Monthly Rider Data")

error = np.random.rand(pyber_ride_df["Month"].count())

pyber_ride_df.plot(kind='barh', x='Month', y="Avg. Fare ($USD)", xerr=error*5,color='red',edgecolor='black',grid=False,
                   ax=ax1, capsize=3,error_kw=dict(ecolor='Black',elinewidth=1),width=0.8)

plt.title="thru plt"
plt.yticks(np.arange(0,51,step=1.0))

pyber_ride_df.plot(kind='bar', x='Month', y="Avg. Fare ($USD)", yerr=error*5,color='skyblue',edgecolor='black',grid=False,
                   ax=ax, capsize=3,error_kw=dict(ecolor='Red',elinewidth=1),width=0.8)
plt.title="thru plt"
plt.xticks(np.arange(0,51,step=2.0))

plt.show()
```
